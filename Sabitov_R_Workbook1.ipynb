## 1.1. –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –º–∞—Ç–µ—Ä–∏–∞–ª ‚Äì –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö

–∏–∑–º–µ–Ω—è–µ–º—ã–µ (mutable) –∏ –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–µ (immutable).
–ù–µ–∏–∑–º–µ–Ω—è–µ–º—ã–µ –æ–±—ä–µ–∫—Ç—ã:

 —Å–∏–º–≤–æ–ª—å–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏ (class 'str'),
 –∫–æ—Ä—Ç–µ–∂–∏ (tuple).
–ò–∑–º–µ–Ω—è–µ–º—ã–µ –æ–±—ä–µ–∫—Ç—ã:
 —Å–ø–∏—Å–∫–∏ (list),

## 1.2. –ü—Ä–∏–º–µ—Ä

–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Ç–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.

```python
x = 3+5.2*7
y = None
z = 'a', 5, 12.345, (2, 'b')
df = [['–ê–Ω—Ç–æ–Ω–æ–≤–∞ –ê–Ω—Ç–æ–Ω–∏–Ω–∞', 34, '–∂'], ['–ë–æ—Ä–∏—Å–æ–≤ –ë–æ—Ä–∏—Å', 26, '–º']]
A = {1, 'title', 2, 'content'}
print(x, '|', type(x), '\n', y, '|', type(y), '\n',
     z, '|', type(z), '\n', df, '|', type(df), '\n', A, '|', type(A))
```

## 1.3. –ó–∞–¥–∞–Ω–∏–µ

–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Ç–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.

```python
x = 5 >= 2
A = {1, 3, 7, 8}
B = {2, 4, 5, 10, 'apple'}
C = A & B
df = '–ê–Ω—Ç–æ–Ω–æ–≤–∞ –ê–Ω—Ç–æ–Ω–∏–Ω–∞', 34, '–∂'
z = 'type'
D = [1, 'title', 2, 'content']
print(x, '|', type(x), '\n', A, '|', type(A), '\n', B, '|', type(B), '\n', C, '|', type(C), '\n',
      df, '|', type(df), '\n', z, '|', type(z), '\n', D, '|', type(D))
```

## 4.3.2 –ó–∞–¥–∞–Ω–∏–µ

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è
–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π 
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π 
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π.

```python
from matplotlib import pyplot as plt
from numpy import sqrt, e, cos, sin, log


def func(x):
    return (sqrt(1 + e ** sqrt(x) + cos(x ** 2)) / abs(1 - (sin(x) ** 3))) + log(abs(2 * x))


answers = [func(x) for x in range(1, 11)]
half_answers = answers[:5]

# –õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
plt.plot(range(1, 11), answers)
plt.grid()

# –¢–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
fig = plt.figure(figsize=(8, 4))
plt.scatter(range(1, 6), half_answers)
plt.grid()
plt.show()
```

## 4.3.3 –ó–∞–¥–∞–Ω–∏–µ

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è |cos(ùë•ùëí^(cos(ùë•)+ln(ùë•+1)))| –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ 
–∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å 
–ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), 
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ 
–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
from numpy import cos, e, log
from matplotlib import pyplot as plt
from scipy.integrate import simps
from numpy import trapz

def func(x):
    return abs(cos(x * e ** (cos(x) + log(x + 1))))


answers = [func(x) for x in range(0, 11)]

area = trapz(answers)
print(area)

# –õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
plt.plot(range(0, 11), answers)
plt.fill_between(range(0, 11), answers)
plt.grid()
plt.show()
```

—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^(ùë•+ùë¶), sin(ùë• + ùë¶), cos(ùë• + ùë¶) , ùë•^y.

```python
from numpy import e, sin, cos

print("–°–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥:\n"
      "1. –°–ª–æ–∂–µ–Ω–∏–µ\n"
      "2. –í—ã—á–∏—Ç–∞–Ω–∏–µ\n"
      "3. –£–º–Ω–æ–∂–µ–Ω–∏–µ\n"
      "4. –î–µ–ª–µ–Ω–∏–µ\n"
      "5. e^(x+y)\n"
      "6. sin(x+y)\n"
      "7. cos(x+y)\n"
      "8. x^y\n"
      "9. –í—ã—Ö–æ–¥")
while True:
    cmd = int(input("–í–≤–µ–¥–∏—Ç–µ –∂–µ–ª–∞–µ–º—É—é –∫–æ–º–∞–Ω–¥—É: "))
    if 0 < cmd <= 9:
        if cmd == 9:
            break
        elif cmd == 1:
            x = float(input("–í–≤–µ–¥–∏—Ç–µ 1-–æ–µ —Å–ª–∞–≥–∞–µ–º–æ–µ: "))
            y = float(input("–í–≤–µ–¥–∏—Ç–µ 2-–æ–µ —Å–ª–∞–≥–∞–µ–º–æ–µ: "))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x + y}")
        elif cmd == 2:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ —É–º–µ–Ω—å—à–∞–µ–º–æ–µ: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ –≤—ã—á–∏—Ç–∞–µ–º–æ–µ: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x - y}")
        elif cmd == 3:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ —É–º–Ω–æ–∂–∞–µ–º–æ–µ: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ –º–Ω–æ–∂–∏—Ç–µ–ª—å: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x * y}")
        elif cmd == 4:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ –¥–µ–ª–∏–º–æ–µ: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ –¥–µ–ª–∏—Ç–µ–ª—å: '))
            if y != 0:
                print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x * y}")
                break
            else:
                print('–û—à–∏–±–∫–∞: –¥–µ–ª–∏—Ç—å –Ω–∞ –Ω–æ–ª—å –Ω–µ–ª—å–∑—è')
        elif cmd == 5:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ x: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ y: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {e ** (x + y)}")
        elif cmd == 6:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ x: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ y: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {sin(x + y)}")
        elif cmd == 7:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ x: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ y: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {cos(x + y)}")
        elif cmd == 8:
            x = float(input('–í–≤–µ–¥–∏—Ç–µ x: '))
            y = float(input('–í–≤–µ–¥–∏—Ç–µ y: '))
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x ** y}")
    else:
        print("–í–≤–µ–¥–∏—Ç–µ –∫–æ–º–∞–Ω–¥—É –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ")
```
